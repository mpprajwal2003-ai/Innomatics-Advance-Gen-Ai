```
def manage_parking_lot(capacity, logs):
    current_count = 0
    peak_usage = 0

    for log in logs:
        if log == "IN":
            current_count += 1
        elif log == "OUT" and current_count > 0:
            current_count -= 1

        # Track peak usage
        if current_count > peak_usage:
            peak_usage = current_count

        # Alert if capacity exceeded
        if current_count > capacity:
            print(" ALERT: Parking lot has exceeded capacity!")

    # Determine parking status
    if current_count >= capacity:
        status = "Full"
    elif current_count >= capacity * 0.8:
        status = "Almost Full"
    else:
        status = "Available"

    print(f"Currently Parked Vehicles: {current_count}")
    print(f"Peak Usage: {peak_usage}")
    print(f"Parking Status: {status}")


# Sample Input
capacity = 50
logs = ["IN", "IN", "IN", "OUT", "IN", "IN", "OUT"]

manage_parking_lot(capacity, logs)
```

```
def estimate_delivery_time(distance, traffic_level, weather):
    # Base speed in km/min (average 30 km/h = 0.5 km/min)
    base_speed = 0.5

    # Traffic delay multipliers
    traffic_delays = {
        "Low":    1.0,
        "Medium": 1.3,
        "High":   1.6
    }

    # Weather delay multipliers
    weather_delays = {
        "Clear": 1.0,
        "Cloudy": 1.1,
        "Rainy":  1.3,
        "Stormy": 1.6
    }

    # Validate inputs
    if traffic_level not in traffic_delays:
        print("Invalid traffic level. Choose: Low, Medium, High")
        return
    if weather not in weather_delays:
        print("Invalid weather. Choose: Clear, Cloudy, Rainy, Stormy")
        return

    # Get multipliers
    traffic_multiplier = traffic_delays[traffic_level]
    weather_multiplier = weather_delays[weather]

    # Calculate base time in minutes
    base_time = distance / base_speed

    # Apply delays
    estimated_time = base_time * traffic_multiplier * weather_multiplier

    # Add fixed preparation time (5 minutes)
    estimated_time += 5

    # Determine delivery status
    if estimated_time <= 30:
        status = " Fast Delivery"
    elif estimated_time <= 50:
        status = " Moderate Delivery"
    else:
        status = " Delayed Delivery"

    print(f"Distance        : {distance} km")
    print(f"Traffic Level   : {traffic_level}")
    print(f"Weather         : {weather}")
    print(f"Delivery Status : {status}")
    print(f"Estimated Delivery Time: {round(estimated_time)} minutes")


# Sample Input
estimate_delivery_time(
    distance=8,
    traffic_level="High",
    weather="Rainy"
)
```

```
def analyze_seat_occupancy(total_seats, booked_seats):
    # Count booked seats
    occupied_seats = booked_seats

    # Calculate occupancy percentage
    occupancy_percentage = (occupied_seats / total_seats) * 100

    # Determine show status
    if occupancy_percentage >= 95:
        status = " Housefull"
        recommendation = "Open additional shows immediately!"
    elif occupancy_percentage >= 80:
        status = " Almost Full"
        recommendation = "Consider opening additional shows."
    elif occupancy_percentage >= 50:
        status = " Moderate Occupancy"
        recommendation = "Monitor bookings."
    else:
        status = " Low Occupancy"
        recommendation = "Promote the show to increase bookings."

    # Display results
    print(f"Total Seats       : {total_seats}")
    print(f"Occupied Seats    : {occupied_seats}")
    print(f"Available Seats   : {total_seats - occupied_seats}")
    print(f"Occupancy         : {occupancy_percentage:.1f}%")
    print(f"Show Status       : {status}")
    print(f"Recommendation    : {recommendation}")

    return {
        "occupancy_percentage": occupancy_percentage,
        "status": status,
        "recommendation": recommendation
    }


# Sample Input
total_seats = 200
booked_seats = 150  # 150 booked seats

analyze_seat_occupancy(total_seats, booked_seats)
```

```
def classify_server_load(cpu_readings):
    # Validate input
    if not cpu_readings:
        print("No CPU readings provided.")
        return

    # Calculate average CPU load
    average_cpu = sum(cpu_readings) / len(cpu_readings)

    # Classify server status
    if average_cpu < 50:
        status = " Normal"
        recommendation = "Server is running smoothly. No action needed."
    elif average_cpu <= 80:
        status = " Warning"
        recommendation = "Monitor closely. Consider load balancing."
    else:
        status = " Critical"
        recommendation = "Immediate action required! Scale up resources."

    # Display results
    print(f"CPU Readings      : {cpu_readings}")
    print(f"Total Readings    : {len(cpu_readings)}")
    print(f"Min CPU Load      : {min(cpu_readings)}%")
    print(f"Max CPU Load      : {max(cpu_readings)}%")
    print(f"Average CPU Load  : {round(average_cpu)}%")
    print(f"Server Status     : {status}")
    print(f"Recommendation    : {recommendation}")

    return {
        "average_cpu": round(average_cpu),
        "status": status,
        "recommendation": recommendation
    }


# Sample Input
cpu_readings = [45, 60, 70, 85, 90]

classify_server_load(cpu_readings)
```

```
def monitor_classroom_resources(resource_usage):
    # Define usage thresholds (in hours)
    thresholds = {
        "Projector": 5,
        "AC":        8,
        "Lights":    6
    }

    overused = []
    energy_alert = False

    print("=" * 50)
    print("   CLASSROOM RESOURCE USAGE REPORT")
    print("=" * 50)

    # Analyze each resource
    for resource, hours in resource_usage.items():
        limit = thresholds.get(resource, 8)  # Default limit: 8 hrs
        usage_percentage = (hours / limit) * 100

        if hours > limit:
            status = " Overused"
            overused.append(resource)
            energy_alert = True
        elif hours == limit:
            status = " At Limit"
        else:
            status = " Normal"

        print(f"\nResource          : {resource}")
        print(f"Hours Used        : {hours} hrs")
        print(f"Allowed Limit     : {limit} hrs")
        print(f"Usage             : {usage_percentage:.1f}%")
        print(f"Status            : {status}")

    # Summary
    print("\n" + "=" * 50)
    print("SUMMARY")
    print("=" * 50)

    if overused:
        print(f"Overused Resources: {', '.join(overused)}")
    else:
        print("Overused Resources: None")

    print(f"Energy Alert      : {' Yes — Reduce usage immediately!' if energy_alert else '✅ No — All resources within limits.'}")

    return {
        "overused_resources": overused,
        "energy_alert": energy_alert
    }


# Sample Input
resource_usage = {
    "Projector": 6,
    "AC":        9,
    "Lights":    4
}

monitor_classroom_resources(resource_usage)
```

```
def manage_event_registration(capacity, registrations):
    # Validate input
    if capacity <= 0:
        print("Invalid capacity. Must be greater than 0.")
        return

    if not registrations:
        print("No registrations provided.")
        return

    # Count total registrations
    total_registrations = len(registrations)

    # Calculate confirmed and waitlisted
    confirmed = min(total_registrations, capacity)
    waitlisted = max(0, total_registrations - capacity)

    # Determine registration status
    if confirmed >= capacity:
        status = " Closed"
        availability = "Event is at full capacity"
    elif confirmed >= capacity * 0.8:
        status = " Almost Full"
        availability = f"Only {capacity - confirmed} seats remaining"
    else:
        status = " Open"
        availability = f"{capacity - confirmed} seats available"

    # Display results
    print("=" * 50)
    print("   EVENT REGISTRATION REPORT")
    print("=" * 50)
    print(f"\nEvent Capacity         : {capacity}")
    print(f"Total Registrations    : {total_registrations}")
    print(f"Confirmed Registrations: {confirmed}")
    print(f"Waitlisted Users       : {waitlisted}")
    print(f"Available Seats        : {capacity - confirmed}")
    print(f"Occupancy Rate         : {(confirmed / capacity) * 100:.1f}%")
    print(f"\nRegistration Status    : {status}")
    print(f"Availability           : {availability}")

    if waitlisted > 0:
        print(f"\n Waitlist Alert: {waitlisted} users are on the waitlist")

    return {
        "confirmed_registrations": confirmed,
        "waitlisted_users": waitlisted,
        "status": status,
        "availability": availability
    }


# Sample Input
capacity = 100
registrations = ["User_" + str(i) for i in range(1, 106)]  # 105 registrations

manage_event_registration(capacity, registrations)
```
