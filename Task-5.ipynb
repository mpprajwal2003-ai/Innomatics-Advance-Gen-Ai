```
def analyze_engagement(likes_list):
    total_likes = 0

    for likes in likes_list:
        total_likes += likes

    if total_likes >= 1000:
        post_status = "Viral Post"
    else:
        post_status = "Normal Engagement"

    print(f"Total Likes: {total_likes}")
    print(f"Post Status: {post_status}")

# Example usage
likes_list = [200, 350, 150, 200, 150]
analyze_engagement(likes_list)
```

```
def check_medicine_stock(medicine_name, stock):
    if stock < 10:
        status = "Low Stock Alert"
    else:
        status = "Stock Sufficient"

    print(f"Medicine: {medicine_name}")
    print(f"Medicine Stock: {stock}")
    print(f"Status: {status}")

# Example usage
check_medicine_stock("Paracetamol", 6)
print("---")
check_medicine_stock("Ibuprofen", 25)
```

```
def check_rainfall(rainfall_list, required_level):
    total_rainfall = 0

    for rain in rainfall_list:
        total_rainfall += rain

    average_rainfall = total_rainfall // len(rainfall_list)

    if average_rainfall >= required_level:
        status = "Adequate Rainfall"
    else:
        status = "Inadequate Rainfall"

    print(f"Average Rainfall: {average_rainfall}")
    print(f"Rainfall Status: {status}")

# Example usage
rainfall_list = [80, 60, 90, 50, 80]
required_level = 70

check_rainfall(rainfall_list, required_level)
```

```
def detect_duplicates(usernames):
    if len(usernames) != len(set(usernames)):
        print("Duplicate Accounts Found: Yes")
    else:
        print("Duplicate Accounts Found: No")

# Example usage
usernames = ["alice", "bob", "alice", "charlie", "dave"]
detect_duplicates(usernames)
```

```
def check_appointment_eligibility(age):
    if age >= 18:
        status = "Eligible"
    else:
        status = "Not Eligible"

    print(f"Patient Age: {age}")
    print(f"Eligibility Status: {status}")

# Example usage
check_appointment_eligibility(21)
```

```
def filter_premium_crops(prices, threshold=2000):
    premium_crops = [price for price in prices if price > threshold]
    print(f"Premium Crops: {premium_crops}")

# Example usage
crop_prices = [1800, 2500, 1500, 3200, 1900, 2100]
filter_premium_crops(crop_prices)
```

```
def check_app_health(error_count):
    if error_count == 0:
        status = "Healthy"
    elif error_count <= 5:
        status = "Minor Issues"
    else:
        status = "Critical Issues"

    print(f"Error Count: {error_count}")
    print(f"System Status: {status}")

# Example usage
check_app_health(7)
```

```
def check_transaction_limit(amount, daily_limit=50000):
    if amount <= daily_limit:
        status = "Approved"
    else:
        status = "Rejected"

    print(f"Transaction Amount: {amount}")
    print(f"Transaction Status: {status}")

# Example usage
check_transaction_limit(60000)
```

```
def check_attendance_eligibility(days_present, total_days):
    attendance_percentage = (days_present / total_days) * 100

    if attendance_percentage >= 75:
        eligibility = "Eligible"
    else:
        eligibility = "Not Eligible"

    print(f"Attendance Percentage: {attendance_percentage:.1f}")
    print(f"Exam Eligibility: {eligibility}")

# Example usage
check_attendance_eligibility(80, 100)
```

```
def calculate_electricity_bill(units):
    if units <= 100:
        bill = units * 3
    elif units <= 200:
        bill = 100 * 3 + (units - 100) * 5
    else:
        bill = 100 * 3 + 100 * 5 + (units - 200) * 7

    # Classify usage
    if bill < 500:
        usage = "Low Usage"
    elif bill <= 1500:
        usage = "Moderate Usage"
    else:
        usage = "High Usage"

    return bill, usage

# Example usage
units_consumed = 250
total_bill, usage_level = calculate_electricity_bill(units_consumed)

print(f"Units Consumed: {units_consumed}")
print(f"Total Bill: ₹{total_bill}")
print(f"Usage Level: {usage_level}")
```
