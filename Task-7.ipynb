```
class SmartLight:
    def __init__(self, name):
        self.name = name
        self.status = "OFF"  # Default state is OFF

    def turn_on(self):
        self.status = "ON"

    def turn_off(self):
        self.status = "OFF"

    def display_status(self):
        print(f"{self.name} is {self.status}")


# --- Main Program ---
light_name = input("Light Name: ")
action = input("Action: ").upper()

light = SmartLight(light_name)

if action == "ON":
    light.turn_on()
elif action == "OFF":
    light.turn_off()
else:
    print("Invalid action! Please enter ON or OFF.")

light.display_status()
```

```
class EmployeeIDCard:
    def __init__(self, name, emp_id, department):
        self.name = name
        self.emp_id = emp_id
        self.department = department

    def display_card(self):
        print("Employee ID Card")
        print(f"Name: {self.name}")
        print(f"ID: {self.emp_id}")
        print(f"Department: {self.department}")


# --- Main Program ---
name = input("Employee Name: ")
emp_id = input("Employee ID: ")
department = input("Department: ")

employee = EmployeeIDCard(name, emp_id, department)
employee.display_card()
```

```
class Contact:
    def __init__(self, name, phone):
        self.name = name
        self.phone = phone

    def display_contact(self):
        print("Contact Saved")
        print(f"Name: {self.name}")
        print(f"Phone: {self.phone}")


# --- Main Program ---
name = input("Contact Name: ")
phone = input("Phone Number: ")

contact = Contact(name, phone)
contact.display_contact()
```

```
class PriceTag:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def display_tag(self):
        print(f"Product: {self.name}")
        print(f"Price: ₹{self.price}")


# --- Main Program ---
product_name = input("Product Name: ")
price = input("Price: ")

tag = PriceTag(product_name, price)
tag.display_tag()
```

```
class MovieRating:
    def __init__(self, name, rating):
        self.name = name
        self.rating = rating

    def display_movie(self):
        print(f"Movie: {self.name}")
        print(f"Rating: {self.rating}/5")


# --- Main Program ---
movie_name = input("Movie Name: ")
rating = input("Rating: ")

movie = MovieRating(movie_name, rating)
movie.display_movie()
```

```
class DeliveryAddress:
    def __init__(self, name, address):
        self.name = name
        self.address = address

    def display_details(self):
        print("Delivery Details")
        print(f"Customer: {self.name}")
        print(f"Address: {self.address}")


# --- Main Program ---
customer_name = input("Customer Name: ")
address = input("Address: ")

delivery = DeliveryAddress(customer_name, address)
delivery.display_details()
```
