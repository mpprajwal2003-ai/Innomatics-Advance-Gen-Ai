```python
def analyze_search_query(query):
    # Convert to lowercase
    query = query.lower()

    # Remove common punctuation
    punctuation = ".,!?;:'\"-"
    for char in punctuation:
        query = query.replace(char, '')

    # Split into keywords
    keywords = query.split()

    # Count frequency of each keyword
    keyword_count = {}
    for keyword in keywords:
        keyword_count[keyword] = keyword_count.get(keyword, 0) + 1

    # Filter keywords that appear more than once
    result = {k: v for k, v in keyword_count.items() if v > 1}

    return result

# Test with the given input
query = "Buy mobile phone buy phone online"
output = analyze_search_query(query)
print(output)
```

```python
def validate_sensor_readings(sensor_readings):
    valid_readings = []

    for hour, reading in enumerate(sensor_readings):
        # Check if reading is even (valid)
        if reading % 2 == 0:
            valid_readings.append((hour, reading))

    return valid_readings

# Test
sensor_readings = [3, 4, 7, 8, 10, 12, 5]
valid_readings = validate_sensor_readings(sensor_readings)

print("Valid Sensor Readings (Hour, Value):")
for hour, value in valid_readings:
    print(f"Hour {hour}: {value}")
print(f"\nList format: {valid_readings}")
```

```python
def analyze_email_domains(emails):
    # Count domains
    domain_count = {}

    for email in emails:
        # Extract domain (part after @)
        domain = email.split('@')[1]
        domain_count[domain] = domain_count.get(domain, 0) + 1

    # Calculate total emails
    total_emails = len(emails)

    # Calculate and display percentages
    for domain, count in domain_count.items():
        percentage = (count / total_emails) * 100
        print(f"{domain}: {percentage:.0f}%")

# Test with the given input
emails = [
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "suresh@gmail.com",
    "meena@yahoo.com"
]

analyze_email_domains(emails)
```

```python
def detect_sales_spikes(sales):
    # Calculate average daily sales
    average_sales = sum(sales) / len(sales)

    # Define threshold for spike (30% above average)
    spike_threshold = average_sales * 1.3

    # Find days with sales more than 30% above average
    spikes = []
    for day, sale in enumerate(sales, start=1):
        if sale > spike_threshold:
            spikes.append((day, sale))

    return spikes

# Test with the given input
sales = [1200, 1500, 900, 2200, 1400, 3000]
spikes = detect_sales_spikes(sales)

print("Sales spikes detected:")
for day, sale in spikes:
    print(f"Day {day}: {sale}")
```

```python
def detect_duplicate_user_ids(user_ids):
    # Count occurrences of each user ID
    user_count = {}

    for user_id in user_ids:
        user_count[user_id] = user_count.get(user_id, 0) + 1

    # Filter and display only duplicates (appearing more than once)
    duplicates = {user_id: count for user_id, count in user_count.items() if count > 1}

    return duplicates

# Test with the given input
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
duplicates = detect_duplicate_user_ids(user_ids)

print("Duplicate User IDs:")
for user_id, count in duplicates.items():
    print(f"{user_id} → {count} times")
```
